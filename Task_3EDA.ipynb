```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})

df.head()
```

```python
df.info()
```

```python

print("Shape of dataset:", df.shape)
print("\nDataset Info:\n")
df.info()

print("\nStatistical Summary:\n")
df.describe()
```

```python
df.hist(figsize=(10, 8), bins=20)
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()
#Insight:
#Numerical features follow near-normal distribution. Petal features show clearer separation.
```

```python
sns.countplot(x='species', data=df)
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
#Insight:All species are equally distributed  no class imbalance.
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop('species', axis=1))
plt.title("Box Plot for Outlier Detection")
plt.show()
#Insight:Minor outliers observed in sepal width, not severe enough to remove.
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df.drop('species', axis=1).corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
#Insight:Petal length and petal width have very strong positive correlation.
```

```python
sns.pairplot(df, hue='species')
plt.show()
#Insight:Setosa is clearly separable. Petal features are best for classification.
```
